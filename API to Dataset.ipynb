```python
import requests
import pandas as pd
```

```python
response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=3545e64349901f61a6e733a3815bd57b&language=en-US&page=1').json()
response
```

```python
response['results']
```

```python
popularity=[]
vote_count=[]
org_lan=[]
title=[]
vote_avg=[]
overview=[]
release_date=[]

for i in response['results']:      #to find the {0}or{1}
 popularity.append(i['popularity'])
 vote_count.append(i['vote_count'])
 org_lan.append(i['original_language'])   
 title.append(i['original_title'])
 vote_avg.append(i['vote_average'])
 overview.append(i['overview'])
 release_date.append(i['release_date'])

d={'title':title,'overview':overview,'release_date':release_date, 'popularity':popularity, 'vote_count':vote_count, 'vote_avg':vote_avg, 'orginal_language':org_lan}
df=pd.DataFrame(d)
df
```
